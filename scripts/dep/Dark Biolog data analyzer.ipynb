```python
# Here we calculate the correct OD with the calculated pathlength

conditions = ['C1', 'C2', 'N']
desktop_path = "/home/asaldivargarci1064/projects/h_lacustris/dark_test"

for condition in conditions:
    file_path = os.path.join(desktop_path, f"{condition}_combined.xlsx")

    if not os.path.exists(file_path):
        print(f"File {file_path} not found. Skipping.")
        continue

    excel_file = pd.ExcelFile(file_path, engine='openpyxl')

    pathlength_sheet = f'{condition}_pathlength_data'
    if pathlength_sheet not in excel_file.sheet_names:
        print(f"Pathlength sheet '{pathlength_sheet}' not found. Skipping {condition}.")
        continue

    pathlength_df = excel_file.parse(pathlength_sheet)
    pathlength_df.columns = pathlength_df.columns.astype(str).str.strip()

    corrected_od_df = pd.DataFrame()

    for sheet in excel_file.sheet_names:
        parts = sheet.strip().split()
        if len(parts) != 3:
            continue

        time, plate, sheet_condition = parts
        if sheet_condition != condition:
            continue

        df = excel_file.parse(sheet)
        df.columns = df.columns.astype(str).str.strip()

        if '680' not in df.columns or 'Wavel.' not in df.columns:
            print(f"Skipping {sheet} - missing required columns.")
            continue

        df = df[df['Wavel.'].astype(str).str.match(r'^[A-H][1-9][0-2]?$')]
        df.reset_index(drop=True, inplace=True)

        pathlength_col = f"{time} {plate} {condition}"
        if pathlength_col not in pathlength_df.columns:
            print(f"Pathlength column '{pathlength_col}' not found. Skipping {sheet}.")
            continue

        pathlength_values = pathlength_df[pathlength_col]

        if len(pathlength_values) != len(df):
            print(f"Row mismatch in {sheet}. Skipping.")
            continue

        corrected_od = df['680'] / pathlength_values
        corrected_od_df[pathlength_col] = corrected_od

        print(f"Processed {sheet}")

    if not corrected_od_df.empty:
        with pd.ExcelWriter(file_path, mode='a', engine='openpyxl', if_sheet_exists='replace') as writer:
            corrected_od_df.to_excel(writer, sheet_name=f"{condition}_corrected_OD", index=False)
        print(f"Corrected OD saved to {file_path}")
    else:
        print(f"No corrected OD data saved for {condition}")
```

```python
# separate corrected OD
import pandas as pd
import os
import re

# Define condition list and plate identifiers
conditions = ['C1', 'C2', 'N']
plates = ['P1', 'P2', 'P3', 'P4']
desktop_path = "/home/asaldivargarci1064/projects/h_lacustris/dark_test/"

for condition in conditions:
    file_path = os.path.join(desktop_path, f"{condition}_combined.xlsx")

    if not os.path.exists(file_path):
        print(f"❌ File {file_path} not found. Skipping.")
        continue

    # Load the corrected OD data
    excel_file = pd.ExcelFile(file_path, engine='openpyxl')
    corrected_sheet_name = f"{condition}_corrected_OD"

    if corrected_sheet_name not in excel_file.sheet_names:
        print(f"❌ Sheet '{corrected_sheet_name}' not found in {file_path}. Skipping.")
        continue

    corrected_df = excel_file.parse(corrected_sheet_name)
    corrected_df.columns = corrected_df.columns.astype(str).str.strip()

    # Output Excel file path
    output_path = os.path.join(desktop_path, f"{condition}_correctOD.xlsx")

    with pd.ExcelWriter(output_path, engine='openpyxl') as writer:
        for plate in plates:
            # Filter and sort columns for this plate
            plate_cols = [col for col in corrected_df.columns if f"{plate} {condition}" in col]
            sorted_plate_cols = sorted(plate_cols, key=lambda x: int(re.search(r'T(\d+)', x).group(1)))

            if not sorted_plate_cols:
                print(f"⚠️ No columns found for {plate} in condition {condition}")
                continue

            plate_df = corrected_df[sorted_plate_cols]
            plate_df.to_excel(writer, sheet_name=plate, index=False)
            print(f"✅ Wrote sheet '{plate}' in {output_path}")

    print(f"✅ Finished writing: {output_path}")
```

```python
# finding natural log

conditions = ['C1', 'C2', 'N'] 
desktop_path = "/home/asaldivargarci1064/projects/h_lacustris/dark_test/"

for condition in conditions:
    file_path = os.path.join(desktop_path, f"{condition}_combined.xlsx")
    
    if not os.path.exists(file_path):
        print(f"File {file_path} not found. Skipping.")
        continue

    excel_file = pd.ExcelFile(file_path, engine='openpyxl')
    corrected_sheet = f"{condition}_corrected_OD"

    if corrected_sheet not in excel_file.sheet_names:
        print(f"Corrected OD sheet '{corrected_sheet}' not found. Skipping {condition}.")
        continue

    corrected_df = excel_file.parse(corrected_sheet)
    corrected_df.columns = corrected_df.columns.astype(str).str.strip()

    log_fc_df = pd.DataFrame()

    for col in corrected_df.columns:
        if not col.startswith("T0 "):
            continue  # We'll use T0 columns as baselines

        # Find all columns matching the same plate/condition
        _, plate, _ = col.split()
        matching_cols = [c for c in corrected_df.columns if c.endswith(f"{plate} {condition}")]
        
        baseline = corrected_df[col]
        if baseline.isna().any() or (baseline <= 0).any():
            print(f"Invalid baseline values in {col}. Skipping plate {plate}.")
            continue

        for mc in matching_cols:
            # Avoid dividing by T0 itself
            if mc == col:
                # Add T0 column directly (no log transformation)
                log_fc_df[mc] = baseline
                continue

            log_fc = np.log(corrected_df[mc] / baseline)
            log_fc = log_fc.apply(lambda x: x if x >= 0 else 0)
            log_fc_df[mc] = log_fc

    if not log_fc_df.empty:
        with pd.ExcelWriter(file_path, mode='a', engine='openpyxl', if_sheet_exists='replace') as writer:
            log_fc_df.to_excel(writer, sheet_name=f"{condition}_logFC_data", index=False)
        print(f" Log fold change saved to {file_path} in sheet '{condition}_logFC_data'")
    else:
        print(f"No valid log fold change data for {condition}")
```

```python
# Separating data to use for gompertz model

conditions = ['C1', 'C2', 'N']
desktop_path = "/home/asaldivargarci1064/projects/h_lacustris/dark_test/"

for condition in conditions:
    file_path = os.path.join(desktop_path, f"{condition}_combined.xlsx")
    
    if not os.path.exists(file_path):
        print(f"File {file_path} not found. Skipping.")
        continue

    excel_file = pd.ExcelFile(file_path, engine='openpyxl')
    logfc_sheet = f"{condition}_logFC_data"

    if logfc_sheet not in excel_file.sheet_names:
        print(f"LogFC sheet '{logfc_sheet}' not found. Skipping {condition}.")
        continue

    logfc_df = excel_file.parse(logfc_sheet)
    logfc_df.columns = logfc_df.columns.astype(str).str.strip()

    # Get all unique plates (P1, P2, ...) for this condition
    plate_names = set()
    for col in logfc_df.columns:
        parts = col.split()
        if len(parts) == 3:
            _, plate, cond = parts
            if cond == condition:
                plate_names.add(plate)

    for plate in sorted(plate_names):
        # Filter columns for this plate and condition
        plate_cols = [col for col in logfc_df.columns if col.endswith(f"{plate} {condition}")]
        plate_df = logfc_df[plate_cols].copy()
        
        # Optional: rename columns to just timepoints (e.g., T0, T1, ...)
        new_col_names = [col.split()[0] for col in plate_cols]
        plate_df.columns = new_col_names

        # Print or store the extracted plate data
        print(f"\n Extracted data for {plate} under {condition}:")
        print(plate_df.head())

        # Optional: Save to new sheet
        with pd.ExcelWriter(file_path, mode='a', engine='openpyxl', if_sheet_exists='replace') as writer:
            sheet_name = f"{condition}_{plate}_logFC"
            plate_df.to_excel(writer, sheet_name=sheet_name, index=False)
        print(f" Saved to sheet '{sheet_name}' in {file_path}")
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score

# --- Gompertz model function ---
def gompertz_model(t, A, u, l):
    return A * np.exp(-np.exp(((u * np.e) / A) * (l - t) + 1))

# --- Timepoint mapping ---
time_mapping = {
    "T0": 0, "T1": 20, "T2": 44, "T3": 70,
    "T4": 95, "T5": 119, "T6": 141, "T7": 164
}
sorted_time_labels = sorted(time_mapping, key=lambda x: time_mapping[x])


# --- Desktop paths ---
desktop_path = "/home/asaldivargarci1064/projects/h_lacustris/dark_test/"
conditions = ['C1', 'C2', 'N']

for condition in conditions:
    file_path = os.path.join(desktop_path, f"{condition}_combined.xlsx")
    if not os.path.exists(file_path):
        print(f"❌ File not found: {file_path}")
        continue

    excel_file = pd.ExcelFile(file_path, engine='openpyxl')
    logfc_sheets = [s for s in excel_file.sheet_names if s.endswith("_logFC")]

    for sheet in logfc_sheets:
        try:
            df = excel_file.parse(sheet)
            available_timepoints = [t for t in sorted_time_labels if t in df.columns]
            df = df[available_timepoints].dropna()
            x_data_local = np.array([time_mapping[t] for t in available_timepoints])
        except Exception as e:
            print(f"⚠️ Skipping sheet {sheet}: {e}")
            continue

        # Set up plot
        fig, axs = plt.subplots(8, 12, figsize=(24, 16))
        axs = axs.flatten()
        fit_results = []

        for i, (well_id, row) in enumerate(df.iterrows()):
            y_data = row.values.astype(float)

            try:
                popt, pcov = curve_fit(
                    gompertz_model, x_data_local, y_data,
                    p0=(2.0, 0.03, 5),
                    bounds=(0, [5, 0.2, 200])
                )
                y_fit = gompertz_model(x_data_local, *popt)
                r2 = r2_score(y_data, y_fit)
            except Exception as e:
                print(e
                popt = [np.nan, np.nan, np.nan]
                y_fit = np.full_like(x_data_local, np.nan)
                r2 = np.nan

            fit_results.append({
                "Well": well_id,
                "Carrying_capacity (A)": popt[0],
                "Growth_rate (u)": popt[1],
                "Lag_time (λ)": popt[2],
                "R2": r2
            })
            if well_id>=50 and well_id<=55:
                print(well_id, popt[0], popt[1], popt[2], r2) 
            # Plot individual well
            if i < len(axs):
                axs[i].plot(x_data_local, y_data, 'o', markersize=3)
                axs[i].plot(x_data_local, y_fit, '-', color='orange')
                axs[i].set_title(well_id, fontsize=6)
                axs[i].set_xlim(0, 170)
                axs[i].set_ylim(0, 3)
                axs[i].tick_params(axis='both', labelsize=5)

        # Remove unused subplots
        for j in range(i+1, len(axs)):
            fig.delaxes(axs[j])

        # Save plot
        plate_name = sheet.replace("_logFC", "")
        plot_filename = f"{plate_name}_gompertz_plot.png"
        plot_path = os.path.join(desktop_path, plot_filename)
        fig.tight_layout()
        fig.savefig(plot_path, dpi=300)
        plt.close(fig)

        # Save fit results to CSV
        fit_df = pd.DataFrame(fit_results)
        csv_filename = f"{plate_name}_Gompertz_fit.csv"
        csv_path = os.path.join(desktop_path, csv_filename)
        fit_df.to_csv(csv_path, index=False)

        print(f"✅ Processed {sheet} — Plot saved to {plot_filename}, CSV saved to {csv_filename}")
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score

# --- Gompertz model function ---
def gompertz_model(t, A, u, l):
    return A * np.exp(-np.exp(((u * np.e) / A) * (l - t) + 1))

# --- Timepoint mapping ---
time_mapping = {
    "T0": 0, "T1": 24, "T2": 45, "T3": 66,
    "T4": 93, "T5": 116, "T6": 142, "T7": 165
}
sorted_time_labels = sorted(time_mapping, key=lambda x: time_mapping[x])
x_data = np.array([time_mapping[t] for t in sorted_time_labels])

# --- Desktop path ---
desktop_path = "/home/asaldivargarci1064/projects/h_lacustris/dark_test/"
conditions = ['C1', 'C2', 'N']

# Process each condition
for condition in conditions:
    file_path = os.path.join(desktop_path, f"{condition}_combined.xlsx")
    if not os.path.exists(file_path):
        print(f"❌ File not found: {file_path}")
        continue

    excel_file = pd.ExcelFile(file_path, engine='openpyxl')
    logfc_sheets = [s for s in excel_file.sheet_names if s.endswith("_logFC")]

    combined_df = []
    for sheet in logfc_sheets:
        try:
            df = excel_file.parse(sheet)
            df = df[sorted_time_labels].dropna()
            df['Well'] = df.index
            combined_df.append(df)
        except Exception as e:
            print(f"⚠️ Skipping sheet {sheet}: {e}")

    if not combined_df:
        continue

    # Average across plates for this condition
    merged_df = pd.concat(combined_df).groupby("Well").mean()
    merged_df = merged_df[sorted_time_labels]

    # Set up plot
    fig, axs = plt.subplots(8, 12, figsize=(24, 16))
    axs = axs.flatten()
    fit_results = []

    for i, (well_id, row) in enumerate(merged_df.iterrows()):
        y_data = row.values.astype(float)

        try:
            popt, pcov = curve_fit(
                gompertz_model, x_data, y_data,
                p0=(2.0, 0.03, 5),
                bounds=(0, [5, 0.2, 200])
            )
            y_fit = gompertz_model(x_data, *popt)
            r2 = r2_score(y_data, y_fit)
        except Exception:
            popt = [np.nan, np.nan, np.nan]
            y_fit = np.full_like(x_data, np.nan)
            r2 = np.nan

        fit_results.append({
            "Well": well_id,
            "Carrying_capacity (A)": popt[0],
            "Growth_rate (u)": popt[1],
            "Lag_time (λ)": popt[2],
            "R2": r2
        })

        if i < len(axs):
            axs[i].plot(x_data, y_data, 'o', markersize=3)
            axs[i].plot(x_data, y_fit, '-', color='orange')
            axs[i].set_title(well_id, fontsize=6)
            axs[i].set_xlim(0, 170)
            axs[i].set_ylim(0, 3)
            axs[i].tick_params(axis='both', labelsize=5)

    for j in range(i+1, len(axs)):
        fig.delaxes(axs[j])

    # Save outputs
    plot_filename = f"{condition}_Average_gompertz_plot.png"
    csv_filename = f"{condition}_Average_Gompertz_fit.csv"

    fig.tight_layout()
    fig.savefig(os.path.join(desktop_path, plot_filename), dpi=300)
    plt.close(fig)

    pd.DataFrame(fit_results).to_csv(os.path.join(desktop_path, csv_filename), index=False)

    print(f"✅ Averaged and processed {condition} — Plot saved to {plot_filename}, CSV saved to {csv_filename}")
```
