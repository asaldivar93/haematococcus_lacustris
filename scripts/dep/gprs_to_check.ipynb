```python
import cobra
import random

import polars as pl
```

```python
file = Path("models/draft/v0.0.2/hlct_haep.xml")
model_h = cobra.io.read_sbml_model(file)
file = Path("models/draft/v0.0.2/hlct_nies.xml")
model_n = cobra.io.read_sbml_model(file)
model_n
```

```python
out_dir = Path("data/1_interim/curation/alexis/")
out_dir.mkdir(exist_ok=True, parents=True)
rxns_to_test = random.sample([rxn.id for rxn in model_n.reactions if bool(rxn.gpr.to_string())], 50)

data = []
for rid in rxns_to_test:
    rxn = model_n.reactions.get_by_id(rid)
    stoi = rxn.build_reaction_string()
    name = rxn.name
    gpr = rxn.gpr.to_string()
    data.append((rid, name, stoi, gpr))

rxns_to_check = pl.DataFrame(data, orient="row", schema=["id", "name", "reaction", "gpr"])
rxns_to_check.write_csv(out_dir / "gprs_to_check_nies.csv")

data = []
for rid in rxns_to_test:
    rxn = model_h.reactions.get_by_id(rid)
    stoi = rxn.build_reaction_string()
    name = rxn.name
    gpr = rxn.gpr.to_string()
    data.append((rid, name, stoi, gpr))

rxns_to_check = pl.DataFrame(data, orient="row", schema=["id", "name", "reaction", "gpr"])
rxns_to_check.write_csv(out_dir / "gprs_to_check_haep.csv")
```
